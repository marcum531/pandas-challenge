```python
#import dependencies
import pandas as pd
```

```python
#upload csv files
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

#read csv files
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#merge the data
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
#Count of schools
Total_Schools = school_data_complete["school_name"].nunique()
Total_Schools
```

```python
#count of students
Total_Students = school_data_complete["student_name"].count()
Total_Students
```

```python
#Budget for all schools
Total_Budget = school_data['budget'].sum()
Total_Budget
```

```python
#Average Math Scores
Average_Math_Scores = school_data_complete['math_score'].mean()
Average_Math_Scores
```

```python
#Average Reading Scores
Average_Reading_Scores = school_data_complete['reading_score'].mean()
Average_Reading_Scores
```

```python
#Passing Math Scores
Passing_Math = school_data_complete.loc[school_data_complete['math_score']>=70]
Passing_Math
```

```python
#Number of students passing math
Passing_Math2 = Passing_Math['student_name'].count()
Passing_Math2
```

```python
#Percent of students passing math
Math_Students = (Passing_Math2/Total_Students)*100
Math_Students
```

```python
#Passing Reading
Passing_Reading = school_data_complete.loc[school_data_complete['reading_score']>=70]
Passing_Reading
```

```python
#Number of students passing reading
Passing_Reading2 = Passing_Reading['student_name'].count()
Passing_Reading2
```

```python
#Percent of students passing reading
Reading_Students = (Passing_Reading2/Total_Students)*100
Reading_Students
```

```python
#Students passing both math and reading
Pass_Reading_Math = school_data_complete.query('reading_score >=70 and math_score >=70')
Pass_Reading_Math
```

```python
#Number of students passing both reading and math
Reading_Math_Students = Pass_Reading_Math['student_name'].count()
Reading_Math_Students
```

```python
#Percent of students passing both math and reading
Percent_Reading_Math_Students = (Reading_Math_Students/Total_Students)*100
Percent_Reading_Math_Students
```

```python
#Dataframe of the school summary
School_data_Summary = pd.DataFrame([[Total_Schools, Total_Students, Total_Budget, Average_Math_Scores, Average_Reading_Scores, Math_Students, Reading_Students, Percent_Reading_Math_Students]], columns=['Total Schools', 'Total Students', 'Total Budget', 'Average Math Scores', 'Average Reading Scores', 'Math Students', 'Reading Students', 'Percent Reading and Math Students'])
School_data_Summary
```

```python
#Schools Summary
Schools_df = school_data_complete.groupby(by = ['school_name','type']).mean()
Schools2_df = Schools_df.drop(columns = ['Student ID', 'School ID'])
Schools2_df

Schools3_df = school_data_complete.set_index(["school_name"])["type"]
Schools3_df
```
