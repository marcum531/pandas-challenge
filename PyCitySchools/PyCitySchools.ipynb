```python
#import dependencies
import pandas as pd
```

```python
#upload csv files
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

#read csv files
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#merge the data
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
Total_Schools = school_data_complete["school_name"].nunique()
Total_Schools
```

```python
Total_Students = school_data_complete["student_name"].count()
Total_Students
```

```python
Total_Budget = school_data['budget'].sum()
Total_Budget
```

```python
Average_Math_Scores = school_data_complete['math_score'].mean()
Average_Math_Scores
```

```python
Average_Reading_Scores = school_data_complete['reading_score'].mean()
Average_Reading_Scores
```

```python
Passing_Math = school_data_complete.loc[school_data_complete['math_score']>=70]
Passing_Math
```

```python
Passing_Math2 = Passing_Math['student_name'].count()
Passing_Math2
```

```python
Math_Students = (Passing_Math2/Total_Students)*100
Math_Students
```

```python
Passing_Reading = school_data_complete.loc[school_data_complete['reading_score']>=70]
Passing_Reading
```

```python
Passing_Reading2 = Passing_Reading['student_name'].count()
Passing_Reading2
```

```python
Reading_Students = (Passing_Reading2/Total_Students)*100
Reading_Students
```

```python
Pass_Reading_Math = school_data_complete.query('reading_score >=70 and math_score >=70')
Pass_Reading_Math
```

```python
Reading_Math_Students = Pass_Reading_Math['student_name'].count()
Reading_Math_Students
```

```python
Percent_Reading_Math_Students = (Reading_Math_Students/Total_Students)*100
Percent_Reading_Math_Students
```

```python

School_data = pd.DataFrame([[Total_Schools, Total_Students, Total_Budget, Average_Math_Scores, Average_Reading_Scores, Math_Students, Reading_Students, Percent_Reading_Math_Students]], columns=['Total_Schools', 'Total_Students', 'Total_Budget', 'Average_Math_Scores', 'Average_Reading_Scores', 'Math_Students', 'Reading_Students', 'Percent_Reading_Math_Students'])
School_data
```
