```python
#import dependencies
import pandas as pd
```

```python
#upload csv files
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

#read csv files
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#merge the data
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
#Count of schools
Total_Schools = school_data_complete["school_name"].nunique()
Total_Schools
```

```python
#count of students
Total_Students = school_data_complete["student_name"].count()
Total_Students
```

```python
#Budget for all schools
Total_Budget = school_data['budget'].sum()
Total_Budget
```

```python
#Average Math Scores
Average_Math_Scores = school_data_complete['math_score'].mean()
Average_Math_Scores
```

```python
#Average Reading Scores
Average_Reading_Scores = school_data_complete['reading_score'].mean()
Average_Reading_Scores
```

```python
#Passing Math Scores
Passing_Math = school_data_complete.loc[school_data_complete['math_score']>=70]
Passing_Math
```

```python
#Number of students passing math
Passing_Math2 = Passing_Math['student_name'].count()
Passing_Math2
```

```python
#Percent of students passing math
Math_Students = (Passing_Math2/Total_Students)*100
Math_Students
```

```python
#Passing Reading
Passing_Reading = school_data_complete.loc[school_data_complete['reading_score']>=70]
Passing_Reading
```

```python
#Number of students passing reading
Passing_Reading2 = Passing_Reading['student_name'].count()
Passing_Reading2
```

```python
#Percent of students passing reading
Reading_Students = (Passing_Reading2/Total_Students)*100
Reading_Students
```

```python
#Students passing both math and reading
Pass_Reading_Math = school_data_complete.query('reading_score >=70 and math_score >=70')
Pass_Reading_Math
```

```python
#Number of students passing both reading and math
Reading_Math_Students = Pass_Reading_Math['student_name'].count()
Reading_Math_Students
```

```python
#Percent of students passing both math and reading
Percent_Reading_Math_Students = (Reading_Math_Students/Total_Students)*100
Percent_Reading_Math_Students
```

```python
#Dataframe of the school summary
School_data_Summary = pd.DataFrame([[Total_Schools, Total_Students, Total_Budget, Average_Math_Scores, Average_Reading_Scores, Math_Students, Reading_Students, Percent_Reading_Math_Students]], columns=['Total Schools', 'Total Students', 'Total Budget', 'Average Math Scores', 'Average Reading Scores', 'Math Students', 'Reading Students', 'Percent Reading and Math Students'])
School_data_Summary
```

```python
#Schools Summary
Schools_df = school_data_complete.groupby(by = ['school_name','type']).mean()
Schools2_df = Schools_df.drop(columns = ['Student ID', 'School ID'])
Schools2_df

Schools3_df = school_data.set_index(["school_name"])["type"]
Schools3_df
```

```python
#Schools Budgets
Schools_Budget = school_data_complete.groupby(["school_name"]).mean()['budget']
Schools_Budget
```

```python
#Schools Budget per Student
Total_Student_Count = school_data_complete["school_name"].value_counts()
Total_Student_Count
```

```python
#Budget per Student
School_Budget_Student = Schools_Budget/Total_Student_Count
School_Budget_Student
```

```python
average_math = school_data_complete.groupby(["school_name"]).mean()['math_score']
average_math
```

```python
#Percent of Students Passing Math
Schools_Passing_Math2 = school_data_complete[(school_data_complete["math_score"]>=70)]
Schools_Passing_Math3 = Schools_Passing_Math2.groupby(["school_name"]).count()["student_name"]/Total_Student_Count*100
Schools_Passing_Math3
```

```python
average_read = school_data_complete.groupby(["school_name"]).mean()['reading_score']
average_read
```

```python
#Percent of Students passing Reading
Schools_Passing_Reading2 = school_data_complete[(school_data_complete["reading_score"]>=70)]
Schools_Passing_Reading3 = Schools_Passing_Reading2.groupby(["school_name"]).count()["student_name"]/Total_Student_Count*100
Schools_Passing_Reading3
```

```python
#Percent of Students passing overall
Schools_Passing_Reading_Math2 = school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70)]
Schools_Passing_Reading_Math3 = Schools_Passing_Reading_Math2.groupby(["school_name"]).count()["student_name"]/Total_Student_Count*100
Schools_Passing_Reading_Math3
```

```python
#Create a dataframe of all lists
Schools_Summary = pd.DataFrame({"School Type":Schools3_df, 
                                "Total Students":Total_Student_Count, 
                                "Total School Budget":Schools_Budget, 
                                "Per Student Budget":School_Budget_Student,
                                "Average Math Score":average_math,
                                "Average Reading Score":average_read,
                                "Passing Math":Schools_Passing_Math3,
                                "Passing Reading":Schools_Passing_Reading3, 
                                "Overall Passing":Schools_Passing_Reading_Math3})
Schools_Summary
```

```python
#Top 5 Performing Schools
Top_Performing = Schools_Summary.sort_values(["Overall Passing"], ascending = False)
Top_Performing.head(5)
```

```python
#Bottom 5 Performing Schools
Low_Performing = Schools_Summary.sort_values(["Overall Passing"], ascending = True)
Low_Performing.head(5)
```

```python
Schools_Summary2 = Schools_Summary.drop(columns = ["School Type", "Total Students", "Total School Budget", "Per Student Budget"])
```

```python
#creating bins and group names to sort data by budget per student
bins = [0, 584, 629, 644, 675]
group_names = ["<$584", "$585-629", "$630-644", "$645 - 675"]
```

```python
#Schools by Spending per student
Schools_Summary2["Spending Ranges(per Student)"] = pd.cut(Schools_Summary["Per Student Budget"], bins, labels=group_names, include_lowest = True)
Schools_Summary3 = Schools_Summary2.groupby("Spending Ranges(per Student)")
Schools_Summary3.mean()
```

```python
#creating bins and group names to sort data by school size
bins1 = [0, 1000, 2000, 5000]
group_names1 = ["Small(<1000)", "Medium(1000-2000)", "Large(2000-5000"]
```

```python
#Schools by School Size
Schools_Summary2["School Size"] = pd.cut(Schools_Summary["Total Students"], bins1, labels=group_names1, include_lowest = True)
Schools_Summary4 = Schools_Summary2.groupby("School Size")
Schools_Summary4.mean()
```

```python
#Schools by type
Schools_Summary5 = Schools_Summary2.groupby['School Type']
Schools_Summary5.mean()
```
